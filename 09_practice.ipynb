# join, splitlines, replace 메서드 실습하기

```python
d1 = '40°' 
m1 = "46'" 
s1 = '52.837"' 
u1 = 'N'

d2 = '73°' 
m2 = "58'" 
s2 = '26.302"' 
u2 = 'W'
```

```python
coords = ' '.join([d1, m1, s1, u1, d2, m2, s2, u2])
print(coords)
```

```python
multi_str="""H: hey hi what's your name
L: my name is seungah
H: nice to meet you
L: see ya"""
```

```python
multi_str_split = multi_str.splitlines()
print(multi_str_split)
```

```python
hey=multi_str_split[::2]
print(hey)
```

```python
hey = multi_str.replace("H: ","").splitlines()[::2]
print(hey)
```

# 문자열 포매팅 실습하기

```python
var = 'flesh wound'
s ="It's just a {}!"

print(s.format(var))
print(s.format('scratch'))
```

```python
s="""what are you doing?
I'm {0} and {1}"""

print(s.format('swimming', 'running'))
```

```python
s="""what are you doing?
I'm {exer1} and {exer2}"""

print(s.format(exer1='swimming', exer2='running'))
```

# 수치값 포매팅 실습

```python
print('some digits of pi: {}'.format(3.141592))
```

```python
print("I'm running {:,}m".format(37469))
```

```python
print('some digits of pi: {0:.4}, {0:.4%}'.format(3.141592))
```

```python
print('my id is {0:05d}'.format(434))
```

# 정규식으로 전화번호 패턴 찾기

```python
import re
tele_num = '1234567890'
```

```python
m = re.match(pattern='\d\d\d\d\d\d\d\d\d\d', string=tele_num) 
print(type(m))
```

```python
print(bool(m))
```

```python
tele_num_spaces='123 456 7890'
```

```python
m = re.match(pattern='\d{10}', string=tele_num_spaces) 
print(m)
```

```python
if m:
    print('match') 
else:
    print('no match')
```

```python
p = '\d{3}\s?\d{3}\s?\d{4}' 
m = re.match(pattern=p, string=tele_num_spaces) 
print(m)
```

```python
tele_num_space_paren_dash = '(123) 456-7890' 
p = '\(?\d{3}\)?\s?\d{3}\s?-?\d{4}' 
m = re.match(pattern=p, string=tele_num_space_paren_dash) 
print(m)
```

```python
cnty_tele_num_space_paren_dash='+1 (123) 456-7890' 
p= '\+?1\s?\(?\d{3}\)?\s?\d{3}\s?-?\d{4}' 
m= re.match(pattern=p, string=cnty_tele_num_space_paren_dash) 
print(m)
```

# compile 메서드로 정규식 메서드 사용하기

```python
p = re.compile('\d{10}') 
s = '1234567890' 
m = p.match(s) 
print(m)
```
